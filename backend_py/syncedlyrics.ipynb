```python
import syncedlyrics
import re
import json

# 1. LRC 형식으로 가사 데이터 가져오기
lyrics_data = syncedlyrics.search("We will rock you movie version", synced_only=True)

# 2. 타임스탬프와 가사를 저장하기 위한 리스트 생성
lyrics_list = []
pattern = r"\[(\d+):(\d+\.\d+)\](.*)"

# 3. 타임스탬프와 가사를 추출하여 'times'와 'lyric'으로 저장
matches = re.findall(pattern, lyrics_data)
for minute, second, lyric in matches:
    # "作词" 또는 "作曲"을 포함하지 않는 가사만 저장
    if "作词" not in lyric and "作曲" not in lyric:
        # 시간을 숫자로 변환 (초 단위로 계산)
        time_in_seconds = int(minute) * 60 + float(second)
        lyrics_list.append({"times": time_in_seconds, "lyric": lyric.strip()})

# 4. 요청된 JSON 구조 생성
response_data = {
    "code": 200,
    "data": {
        "lyrics": lyrics_list
    }
}

# 5. JSON 문자열로 변환
json_result = json.dumps(response_data, ensure_ascii=False, indent=4)

# 6. 결과 출력
print(json_result)
```
